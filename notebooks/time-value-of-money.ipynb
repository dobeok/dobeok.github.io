## Time value of money

Assume that $100 is deposited in a bank today with an annual interest rate of 10\%, what is the value of the reposit one year later?

```python
# pv = present value
# fv = future value
# r = simple interest rate
# fv = pv + pv * r

pv = 100
r = .1
fv_year1 = pv + pv * r
fv_year1
```

Assume that $100 is keptin the bank for another year with the same annual interest rate. What will be the future value at the end of year two?

```python
# using the same logic above
# we treat the future value at the end of year 1 as the present value at the end of year 1

fv_year2 = pv + pv * r + (pv + pv * r) * r
fv_year2
```

The general formular for simple interest rate is $fv = pv*(1+r)^n$ where $n$ is the number of periods
