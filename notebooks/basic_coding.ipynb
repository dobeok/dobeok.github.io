### FizzBuzz

```python
# fizz buzz
def fb(n: int):
    for i in range(1, n+1):
        
        out = ''
        
        if i%3==0:
            out+='Fizz'
            
        if i%5==0:
            out+='Buzz'
        
        if i%3 and i%5:
            out=str(i)
            
        print(out)

fb(16)
```

### Check string palindrome

```python
def check_palindrome(text: str):
    return text, text == text[::-1]

print(check_palindrome('abcd'))
print(check_palindrome('abcdcba'))
print(check_palindrome('xyyx'))
print(check_palindrome('xyx'))
```

### Mean, Median, Variane of list

```python
def mean_of_list(num: list):
    return sum(num) / len(num)

mean_of_list([1,2,3,4,5,6])
```

```python
def var_of_list(num: list):
    """
    calculating as population variance
    """
    mean = mean_of_list(num)

    return sum([(el - mean)**2 for el in num]) / len(num)

var_of_list([1,2,3,4,5,6])
```

```python
def median_of_list(num: list):
    sorted_num = sorted(num)
    
    # odd numbers of elements     
    if len(num) % 2:
        median = sorted_num[len(num)//2]
    else:
        median = (sorted_num[len(num)//2-1] + sorted_num[len(num)//2]) / 2
    
    return median

print(median_of_list([4,3,2,1,0]))
print(median_of_list([9,10,10,30]))
print(median_of_list([-1, 10000, 3, 4, 2, 1, 5, 5, 100]))
    
```
