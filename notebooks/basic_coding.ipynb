### FizzBuzz

```
# fizz buzz
def fb(n: int):
    for i in range(1, n+1):
        
        out = ''
        
        if i%3==0:
            out+='Fizz'
            
        if i%5==0:
            out+='Buzz'
        
        if i%3 and i%5:
            out=str(i)
            
        print(out)

fb(16)
```

### Check palindrome

```
def check_palindrome(text: str):
    return text, text == text[::-1]

print(check_palindrome('abcd'))
print(check_palindrome('abcdcba'))
print(check_palindrome('xyyx'))
print(check_palindrome('xyx'))
```

### Mean and Variane of list

```
def mean_of_list(num: list):
    return sum(num) / len(num)

mean_of_list([1,2,3,4,5,6])
```

```
def var_of_list(num: list):
    mean = mean_of_list(num)

    return sum([(el - mean)**2 for el in num]) / len(num)

var_of_list([1,2,3,4,5,6])
```
